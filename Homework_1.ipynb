```python
import math as m
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
%matplotlib notebook
```

```python
def f(n, lamdba):
    lst = [0]
    
    for k, nk in enumerate(n, start=1):
        if k > 1:
            lst.append(1/2 * np.dot(n[:k-1], n[k-2::-1]) - (1 + lamdba) * nk * np.sum(n))
              
    return np.array(lst)
```

```python
def Euler(t0, b, h, n0, f, lamdba):
    max_iter = m.ceil((b-t0)/h)
    
    n_lst = [n0]
    t_lst = [t0]
    
    t = t0
    n = n0

    for i in range(max_iter):
        t += h
        n = n + f(n, lamdba)*h
          
        n_lst.append(n)
        t_lst.append(t)
    
    return np.array(n_lst), np.array(t_lst)
```

```python
N = 100
n0 = np.zeros(N)
n0[0] = 1

nk, t = Euler(0, 10, 0.01, n0, f, 0.3)
```

```python
plt.figure("n2(t)")
plt.plot(t, nk[:, 1])
plt.xlabel("t")
plt.ylabel("n2");
```

```python
plt.figure("n4(t)")
plt.plot(t, nk[:, 4])
plt.xlabel("t")
plt.ylabel("n4");
```

```python
plt.figure("nk(k)")
plt.plot(np.arange(1,101), nk[-1, :])
plt.xlabel("k")
plt.ylabel("nk");
```

```python
plt.figure("log nk(k)")
plt.plot(np.arange(1,101), nk[-1, :])
plt.xlabel("k")
plt.ylabel("nk")
plt.yscale('log');
```
